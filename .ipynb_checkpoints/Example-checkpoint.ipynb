```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import torch, scipy.io
import main 
import torch.nn.functional as F
import copy
```

# Input

### Load training data

```python
input_data = np.load('example_input.npy').item()
spk = input_data['spk']
h = input_data['h'] # initial guess for 2d reconstruction filter (based on 1d reconstruction filter)
env = input_data['env']
print(f'Spk dimensions {spk.shape}')
print(f'Initial 2d recon filter h dimensions {h.shape}')
print(f'Env dimensions {env.shape}')

h_before = copy.deepcopy(h) # copy initial h to compare after training
```

### Visualize inputs

```python
plt.figure(figsize=[8,4],dpi=300)
n_t = spk.shape[2]
n_h = h.shape[2]
n_f = h.shape[1]
n_output = n_t - n_h + 1

plt.subplot(2,1,1)
plt.plot(env, 'red')
plt.title('Envelope')
plt.xlabel('Time step')
plt.xlim([0, n_t])

plt.subplot(2,1,2)
plt.imshow(np.squeeze(spk), aspect='auto', cmap='binary')
plt.title('Spikes')
plt.ylabel('Frequency channel')
plt.xlabel('Time step')
plt.tight_layout()
plt.xlim([0, n_t])

# plt.subplot(3,2,5)
# plt.imshow(np.squeeze(h), aspect='auto', cmap='Blues')
# plt.title('Reconstruction filter')
# plt.xlim([0, n_h])
# plt.ylim([0, n_f])
# plt.xlabel('Time step')
# plt.ylabel('Frequency channel')
```

### Convert inputs into pytorch tensors

```python
device = torch.device('cpu')

env = torch.from_numpy(env)
spk = torch.from_numpy(spk)
h = torch.from_numpy(h)
h.requires_grad = True
```

# Find optimal filter

### Gradient descent

```python
from importlib import reload  # Python 3.4+ only.
import main 
reload(main)
from main import TwoDimReconstruction
recon_model = TwoDimReconstruction(10, learning_rate=.0001)
h, est_env, loss_history = recon_model.fit(spk, h, env)
```

### Plot before and after

Calculate envelopes before and after

```python
est_env_before = F.conv1d(spk,torch.from_numpy(h_before)) # estimated enveloe
loss_before = (est_env_before - env[:n_output]).pow(2).sum() # loss function
print(f'Loss before training {loss_before:.2f}')

est_env_after = F.conv1d(spk,h) # estimated enveloe
loss_after = (est_env_after - env[:n_output]).pow(2).sum() # loss function
print(f'Loss after training {loss_after:.2f}')
```

Plot

```python
import matplotlib as mpl
mpl.rcParams["font.size"] = 8
mpl.rcParams["axes.titlesize"] = 12
plt.figure(figsize=[10,5],dpi=300)

plt.subplot(2,3,1)
plt.plot(loss_history)
plt.title('Loss')
plt.xlabel('Run')

plt.subplot(2,3,2)
plt.imshow(np.squeeze(h_before), aspect='auto', cmap='Blues')
plt.title('Reconstruction filter: before')
plt.xlim([0, n_h])
plt.ylim([0, n_f])

plt.subplot(2,3,3)
plt.imshow(np.squeeze(h.detach().numpy()), aspect='auto', cmap='Blues')
plt.title('Reconstruction filter: after')
plt.xlim([0, n_h])
plt.ylim([0, n_f])

plt.subplot(2,1,2)
plt.plot(env.numpy(), alpha=.5)
plt.plot(est_env_before[0][0].numpy(), alpha=.5)
plt.plot(est_env_after[0,0,:].detach().numpy(), alpha=.5)
plt.legend(['Original', f'Reconstruction before: loss={loss_before:.2f}', 
            f'Reconstruction after: loss={loss_after:.2f}'])
plt.title('Envelope')
plt.xlabel('Time')
plt.xlim([0, n_output])

plt.tight_layout()
```
